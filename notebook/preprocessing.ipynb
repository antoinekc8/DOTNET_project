```python
# check if a google link is in df_inaccessible_od from df_destination_accessibility and df_origin_accessibility

for i in range(len(df_origin_accessibility)):
    check = df_origin_accessibility['google_maps_link'][i]
    for link in df_inaccessible_od['google_maps_http_link']:
        if check == link:
            print("Found")
        
for i in range(len(df_destination_accessibility)):
    check = df_destination_accessibility['google_maps_link'][i]
    for link in df_inaccessible_od['google_maps_http_link']:
        if check == link:
            print("Found")

for i in range(len(df_google_maps_od_distance)):
    check = df_google_maps_od_distance['google_maps_http_link'][i]
    for link in df_inaccessible_od['google_maps_http_link']:
        if check == link:
            print("Found")       



# clean the problematic distances from google maps od distances

# Nombre de lignes avant filtrage
n_before = len(df_google_maps_od_distance)

# Création des paires OD problématiques
problematic_pairs = set(
    zip(
        df_problematic_od_distances['o_zone_id'],
        df_problematic_od_distances['d_zone_id']
    )
)

# Création du mask
mask = ~df_google_maps_od_distance[['o_zone_id', 'd_zone_id']].apply(
    tuple, axis=1
).isin(problematic_pairs)

# Application du filtre
df_google_maps_od_distance_clean = df_google_maps_od_distance[mask]

# Nombre de lignes après filtrage
n_after = len(df_google_maps_od_distance_clean)

# Affichage
print(f"{n_before - n_after} trips have been removed from google maps od distances")
print(f"{n_after} trips remaining in google maps od distances")




# to clean od performance from problematic distances (46 mins), to uncomment

problematic_pairs = set(
    zip(
        df_problematic_od_distances['o_zone_id'],
        df_problematic_od_distances['d_zone_id']
    )
)

mask = ~df_od_performance[['o_zone_id', 'd_zone_id']].apply(
    tuple, axis=1
).isin(problematic_pairs)

df_od_performance_clean = df_od_performance[mask]

n_before = len(df_od_performance)

df_od_performance_clean = df_od_performance[mask]

n_after = len(df_od_performance_clean)

print(f"{n_before - n_after} trips have been removed from od performance")
print(f"{n_after} trips remaining in od performance")
```

Cleaning:

```python
len_link_before = len(df_link)

for links in df_problem_volume_links.itertuples():
    df_link_performance = df_link_performance.drop(links.Index)
    df_link = df_link.drop(links.Index)
    
print(f'We have removed {len_link_before - len(df_link)} problematic links from link and link performance dataframes')
```

```python
# A link is a connector if it has length 0, time 0, or specific link_type (e.g., 0)
is_connector = (df_link['length'] <= 0) | (df_link['vdf_fftt'] <= 0) | (df_link['link_type'] == 0)

print(f"Total links: {len(df_link)}")
print(f"Identified connectors: {is_connector.sum()}")
print(f"Identified real roads: {(~is_connector).sum()}")


# Parameter Configuration for CONNECTORS
# Connectors must be "transparent": minimum time, no congestion
df_link.loc[is_connector, 'vdf_fftt'] = 0.1    # 6 seconds fixed time
df_link.loc[is_connector, 'length'] = 0.1      # Minimum distance
df_link.loc[is_connector, 'capacity'] = 99999  # Virtually infinite capacity
df_link.loc[is_connector, 'vdf_alpha'] = 0.0   # No time increase with volume
df_link.loc[is_connector, 'vdf_beta'] = 1.0

# Parameter Configuration for REAL ROADS
# Apply standard BPR parameters to model congestion
real_roads = ~is_connector
df_link.loc[real_roads, 'vdf_alpha'] = df_link.loc[real_roads, 'vdf_alpha'].fillna(0.15)
df_link.loc[real_roads, 'vdf_beta'] = df_link.loc[real_roads, 'vdf_beta'].fillna(4.0)

# Filling missing capacities with median (excluding 99999 from connectors)
median_cap = df_link.loc[real_roads & (df_link['capacity'] < 90000), 'capacity'].median()
df_link.loc[real_roads, 'capacity'] = df_link.loc[real_roads, 'capacity'].fillna(median_cap)

# Node and Demand Alignment (Integer Casting)
# Transform zone_id to integer for perfect matching with demand
df_node['zone_id'] = pd.to_numeric(df_node['zone_id'], errors='coerce')

# Create a cleaned version of nodes for export
nodes_cleaned = df_node.copy()
# Handle NaNs before casting
nodes_cleaned['zone_id'] = nodes_cleaned['zone_id'].fillna(-1).astype(int)

# Filter demand keeping only valid zones
valid_zones = nodes_cleaned[nodes_cleaned['zone_id'] > 0]['zone_id'].unique()
demand_cleaned = df_demand[df_demand['o_zone_id'].isin(valid_zones) & df_demand['d_zone_id'].isin(valid_zones)].copy()
```
